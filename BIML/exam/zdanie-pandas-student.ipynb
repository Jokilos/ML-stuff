### 1. Load the data (1p)

```python
data = pd.read_csv('./evaluation_data.csv')
df = data
df
```

### 2. What is the run id of a non failed experiment that achieved the highest accuracy? (3p)

```python
non_failed = df[df['failed'] == False]
id = non_failed.sort_values('accuracy').tail(1).iloc[0,0]

best_run_id = id
print(best_run_id)
```

### 3. Please select rows from oryginal dataset that only contain models that have full set of results (evaluations on all datasets have succeeded). (3p)

```python
gb = df.groupby('model_id')
full = gb['failed'].aggregate('sum') == 0
indexes = df['model_id'].map(lambda x : full[x])

fully_evaluated_data = df[indexes]
print(fully_evaluated_data)
```

### 4. Using data obtained in the previous step answer the following question. What is the id of the best performing model (with the highest accuracy across datasets)? (3p)

```python
#I understood that we only inspect non-failing models
gb1 = fully_evaluated_data.groupby('model_id')
best = gb1['accuracy'].aggregate('mean').tail(1)
id = best.index[0]

best_model_id = id
print(best_model_id)
```
