## 1. Data Loading and Visualization (1p)

```python
# Import necessary libraries
from typing import Tuple
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# For inline plotting in Jupyter notebooks
%matplotlib inline

def load_data(file_path: str) -> Tuple[np.ndarray, np.ndarray]:
    """Load dataset from a CSV file.

    Args:
        file_path: Path to the CSV file.

    Returns:
        A tuple containing features X and target y.
    """
    data: pd.DataFrame = pd.read_csv(file_path)
    X: np.ndarray = data[['X']].values  # Ensure X is a 2D array
    y: np.ndarray = data['y'].values
    return X, y

#TODO: Load the dataset 'data.csv'

X, y = load_data('data.csv')

def plot_data(X: np.ndarray, y: np.ndarray) -> None:
    """Plot the data points.

    Args:
        X: Feature array.
        y: Target array.
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(X, y, color='blue', label='Data Points')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Scatter Plot of y vs X')
    plt.legend()
    plt.show()

# Plot y versus X
plot_data(X, y)
```

## 2. Model Fitting (3p)

### 2.1 Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

def train_linear_regression(X: np.ndarray, y: np.ndarray) -> LinearRegression:
    """Train a Linear Regression model.

    Args:
        X: Feature array.
        y: Target array.

    Returns:
        Trained LinearRegression model.
    """
    linear_model: LinearRegression = LinearRegression().fit(X,y)
    return linear_model
    # TODO: Fit (and return) the model to X and y

# Train the linear regression model
linear_model = train_linear_regression(X, y)
```

### 2.2 Polynomial Regression Model (Degree 3)

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def train_polynomial_regression(X: np.ndarray, y: np.ndarray, degree: int) -> make_pipeline:
    """Train a Polynomial Regression model.

    Args:
        X: Feature array.
        y: Target array.
        degree: Degree of the polynomial features.

    Returns:
        Trained polynomial regression pipeline.
    """
    # TODO: Create a pipeline with PolynomialFeatures and LinearRegression
    polynomial_model: PolynomialFeatures = PolynomialFeatures(degree).fit(X,y)
    # TODO: Fit the polynomial model to X and y

    return polynomial_model

degree: int = 3
# Train the polynomial regression model
polynomial_model = train_polynomial_regression(X, y, degree)
type(polynomial_model)
```
