```python
import subprocess
import numpy as np

n = 10
props = np.linspace(0.018, 0.03, n)
ders = np.linspace(0.18, 0.3, n)
xprops, yders = np.meshgrid(props, ders)
res = np.zeros((n, n))
script_name = "runnable.py"

for x in range(n):
    for y in range(n):
        prop = xprops[x][y]
        der = yders[x][y]
        print(f"{prop=} {der=}")

        for _ in range(3): 
            result = subprocess.run(
                ["python", script_name, str(prop), str(der)],
                text = True, 
                capture_output = True,
            )
            
            err = float(result.stdout)
            res[x][y] += err
            print(err)
        
        print("")

print(res)
print(np.argmin(res))
print(np.min(res))
```
